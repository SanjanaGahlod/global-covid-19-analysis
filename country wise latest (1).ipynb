```python
data.plot(kind = 'box' , figsize =(25,15),subplots=True)
```

```python
Death_Case = data[['Country/Region','Deaths']].sort_values(by=['Deaths'],ascending=False).head(10)
Death_Case
```

```python
plt.style.use('seaborn-poster')
plt.figure(figsize=(15,7))
plt.title('Top 10 Recovered Countries')
plt.xlabel('Number of Recoveries')
top_recovered = data.groupby('Country/Region')['Recovered'].sum().nlargest(10)
top_recovered.plot(kind='barh',color='y')
for index, value in enumerate(top_recovered):
    plt.text(value, index, str(value))  # Add text labels

plt.show()

```

```python
Active_Cases = data[['Country/Region','Active']].sort_values(by=['Active'],ascending=False).head(15)
Active_Cases
```

```python
sns.barplot(data = Active_Cases , x = 'Active' , y = 'Country/Region')
plt.title('Active Cases')
```
